2.  What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot

A scatter plot is a type of data visualization that displays the relationship between two variables. It is a graph in which each individual data point is represented as a dot or point on the plot, with one variable on the horizontal (x) axis and the other variable on the vertical (y) axis.

Scatter plots are commonly used to identify trends and relationships between variables, such as the correlation between two variables. They can also be used to identify clusters or groups in the data.

Scatter plots are particularly useful when working with numerical data, as they allow analysts to quickly see patterns and relationships between variables. They can also be used to identify outliers or anomalies in the data.

In a scatter plot, the position of each data point represents the value of the two variables being compared. The closer the points are to a diagonal line, the stronger the relationship between the variables. If the points are widely dispersed, it suggests a weak or non-existent relationship.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.title("RANDOM Ploting")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

3. Why is the subplot() function used?

The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows you to arrange several plots in a grid, with each plot sharing the same axis or having its own individual axis.

The subplot() function takes three arguments: nrows, ncols, and index.

nrows and ncols specify the number of rows and columns in the grid of plots, respectively. The index argument is used to specify the location of the subplot within the grid, and it ranges from 1 to nrows * ncols.

```python
x=np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
y1 = np.array([50, 20, 40, 20, 60, 70])
y2 = np.array([10, 20, 30, 40, 50, 60])
y3= np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
```

4.  What is a bar plot? Why is it used?

A bar plot is a type of chart or graph that displays categorical data with rectangular bars. The height or length of each bar is proportional to the value or frequency of the category it represents. Bar plots are commonly used to compare the values of different categories, or to display how a categorical variable changes over time.

Bar plots are useful because they make it easy to compare the relative sizes of different categories. They can be used to visualize a variety of data, such as counts, proportions, averages, and percentages. Bar plots are also easy to create and interpret, making them a popular choice for data visualization.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.title("profit v/s compny")
plt.xlabel('company name')
plt.ylabel("profit")
           
        
```

5. What is a box plot? Why is it used? 

A box plot is a type of chart or graph that displays the distribution of a dataset through their quartiles. It consists of a box, which represents the middle 50% of the data, and "whiskers" which extend out from the box and represent the minimum and maximum values of the data. Box plots can also display "outliers", which are data points that are further away from the other data points than expected.

Box plots are used to provide a visual summary of the distribution of a dataset, and to compare the distribution of multiple datasets. They are especially useful when dealing with large datasets or datasets that have many outliers. Box plots are also commonly used in statistical analysis and research to identify potential skewness, outliers and to identify if the data is normally distributed.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title('Box Plot Comparison')
plt.xlabel('Data')
plt.ylabel('Value')
```
